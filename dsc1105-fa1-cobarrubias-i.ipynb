```r
library(tidyverse)
library(ggplot2)
```

```r
data = read_csv("/kaggle/input/cytof-one-experiment-csv/cytof_one_experiment.csv")
```

```r
hist(data$NKp30, xlab = "NKp30",
     col = "black", border = "black")
```

```r
ggplot(data, aes(x = NKp30)) + geom_density()
```

```r
ggplot(data, aes(x = sort(NKp30), y = sort(KIR3DL1))) +
  geom_point() +
  geom_abline(slope = 1, intercept = 0, color = "red") + 
  xlab("NKp30 Quantiles") + 
  ylab("KIR3DL1 Quantiles") +
  ggtitle("Q-Q Plot: NKp30 vs KIR3DL1")
```

The plotted dots form a curved line which shows the 2 columns having different skewness. On the end, however, the dots follow the slope of the diagonal but is shifted. This could mean that part shows similar behavior of the data but with different intensity of the values.
